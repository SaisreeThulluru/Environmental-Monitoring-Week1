This notebook contains basic data analysis and visualization for pollution monitoring.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("dataset.csv")
df.head()
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['PM2.5'], marker='o')
plt.xticks(rotation=45)
plt.title('PM2.5 Levels Over Time')
plt.xlabel('Date')
plt.ylabel('PM2.5 (¬µg/m¬≥)')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['NO2'], color='orange', marker='s')
plt.xticks(rotation=45)
plt.title('NO‚ÇÇ Levels Over Time')
plt.xlabel('Date')
plt.ylabel('NO‚ÇÇ (ppb)')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['CO'], color='green', marker='^')
plt.xticks(rotation=45)
plt.title('CO Levels Over Time')
plt.xlabel('Date')
plt.ylabel('CO (mg/m¬≥)')
plt.tight_layout()
plt.show()
```

### ‚úÖ Conclusion
- Pollution levels fluctuate daily.  
- PM2.5 levels are relatively higher compared to NO‚ÇÇ and CO.  
- Further steps can include predicting pollution levels using ML models.
